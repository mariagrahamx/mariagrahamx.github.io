```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/project/glasgow_simd_2016_vs_2020_x_distance_to_village.csv"
simd_medians_long = pd.read_csv(url)
simd_medians_long
```

```python
# split years
med_2016 = simd_medians_long[simd_medians_long["year"] == 2016]
med_2020 = simd_medians_long[simd_medians_long["year"] == 2020]
```

```python
# merge on columns
change_df = med_2016.merge(
    med_2020,
    on=["village_band", "domain"],
    suffixes=("_2016", "_2020")
)
```

```python
# comparability
change_df["delta_median_rank"] = change_df["median_rank_2020"] - change_df["median_rank_2016"]
```
