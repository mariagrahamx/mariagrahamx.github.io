```python
import pandas as pd
from pyproj import Transformer
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/wk7/wk7_derelict_land_scotland_points_raw.csv"

data1 = pd.read_csv(url)
data1.head(10)
```

```python
# set the column names to the values of the seventh row
data1.columns = data1.iloc[6]
data1
```

```python
# drop the first 8 rows
data2 = data1.iloc[7:]
data2
```

```python
# delete columns I don't want
data3 = data2.drop(columns=['Owner 1', 'Owner 2','Previous Use of Site', 'Development Potential', 'Site Code', 'Location of Site'])
data3
```

```python
data4 = data3.rename(columns={'Site Name (If Supplied)': 'Site Name', 'Address (If Supplied)': 'Address', 'Site Size (Hectares)':'Site Size'})
data4
```

```python
# convert east/north to lat/long
transformer = Transformer.from_crs("EPSG:27700", "EPSG:4326", always_xy=True)

lons = []
lats = []

for x, y in zip(data4["East"], data4["North"]):
    lon, lat = transformer.transform(x, y)
    lons.append(lon)
    lats.append(lat)

data4["lon"] = lons
data4["lat"] = lats
```

```python
# copy new data frame to data5
data5 = data4.copy()
data5
```

```python
data5.columns = data5.columns.str.strip()
df_glasgow = data5[data5["Planning Authority"] == "Glasgow City"]

df_glasgow
```

```python
import re

def extract_start_year(period):
    if pd.isna(period):
        return None

    period = str(period).lower()

    # case 1: "1980 or earlier"
    if "or earlier" in period:
        return int(re.search(r"\d{4}", period).group())

    # case 2: range like "1981-1985"
    if "-" in period:
        return int(period.split("-")[0])

    # case 3: single year
    match = re.search(r"\d{4}", period)
    if match:
        return int(match.group())

    return None

df_glasgow2 = df_glasgow.copy()

df_glasgow2["Start Year"] = df_glasgow2["Period when site became Vacant or Derelict"].apply(extract_start_year).astype("Int64")

df_glasgow2

```

```python
df_glasgow2.to_csv ("glasgow_derelict_sites_clean.csv", index=False)
```
