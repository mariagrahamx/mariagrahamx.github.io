```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/Capstone/World%20Bank%20Poverty%20Index%201963%20to%202024.csv"

df = pd.read_csv(url)
df.head(15)
```

```python
# set the column names to the values of the sixth row

df.columns = df.iloc[5]
df.head(15)
```

```python
# drop the first 5 rows now that it has been set as the column names

df2 = df.drop(df.index[0:6])
df2.head(15)
```

```python
# check structure

print('Current Shape:', df2.shape)
print('Columns:', df2.columns.tolist())
```

```python
# drop columns we don't want

# The last column in df2.columns is np.float64(nan), which can cause issues with direct dropping.
# Access it by its position to ensure the exact object is passed to drop.
nan_column_name = df2.columns[-1]

df3 = df2.drop(columns=['Indicator Code', 2020.0, 2021.0, 2022.0, 2023.0, 2024.0, 2025.0, nan_column_name])
df3
```

```python
print(df3['Indicator Name'].unique())
```

```python
# select indicators we want

variable_to_keep = 'Poverty headcount ratio at $3.00 a day (2021 PPP) (% of population)'

df_poverty = df3[df3['Indicator Name'] == variable_to_keep].copy()

print(f"Filtered to {len(df_poverty)} rows")
```

```python
df_poverty.head(40)
```

```python
# clean name of indicator

variable_mapping = {
    'Poverty headcount ratio at $3.00 a day (2021 PPP) (% of population)': 'poverty_headcount_3dollar'}

df_poverty['variable_name'] = df_poverty['Indicator Name'].map(variable_mapping)

df_poverty
```

```python
# get year columns (they're floats)

year_cols = [col for col in df_poverty.columns if isinstance(col, float) and not pd.isna(col)]

print(f"Years: {min(year_cols)} to {max(year_cols)}")

# melt to long format
df_long = df_poverty.melt(
    id_vars=['Country Name', 'Country Code', 'variable_name'],
    value_vars=year_cols,
    var_name='year',
    value_name='value'
)

# convert year to int
df_long['year'] = df_long['year'].astype(int)
df_long['value'] = pd.to_numeric(df_long['value'], errors='coerce')

print(f"\nLong format shape: {df_long.shape}")
print(df_long.head())
```

```python
# check how much actual data we have (not NaN)

print(f"Total rows: {len(df_long)}")
print(f"Non-missing poverty data: {df_long['value'].notna().sum()}")
print(f"Missing: {df_long['value'].isna().sum()}")
```

```python
df_long
```

```python
# pivot to wide format

df_wide = df_long.pivot_table(
    index=['Country Name', 'Country Code', 'year'],  # one row per country-year
    columns='variable_name',            # each indicator becomes a column
    values='value',
    aggfunc='first'  # in case of any remaining duplicates
).reset_index()

print(f"\nWide format shape: {df_wide.shape}")
print(f"Columns: {df_wide.columns.tolist()}")
```

```python
df_wide
```

```python
df_poverty_clean = df_wide.copy()

# Rename columns for consistency
df_poverty_clean = df_poverty_clean.rename(columns={
    'Country Code': 'ISO',
    'Country Name': 'Country'
})

# Filter to 1980-2019
df_poverty_clean = df_poverty_clean[(df_poverty_clean['year'] >= 1980) &
                                     (df_poverty_clean['year'] <= 2019)]

print(f"Cleaned poverty data: {df_poverty_clean.shape}")
print(df_poverty_clean.head())
```
