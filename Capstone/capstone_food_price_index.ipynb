```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/Capstone/food_price_index_nominal_real_1960-2026.csv"

df = pd.read_csv(url)
df.head(15)
```

```python
# set the column names to the values of the second row

df.columns = df.iloc[1]
df.head(15)
```

```python
# drop the first 3 rows

df2 = df.drop(df.index[0:3])
df2.head(15)
```

```python
# check structure

print('Current Shape:', df2.shape)
print('Columns:', df2.columns.tolist())
```

```python
# drop columns we don't want

# Identify columns that are NaN
nan_cols = [col for col in df2.columns if pd.isna(col)]

# Combine 'Real' with the identified NaN columns
columns_to_drop = ['Real'] + nan_cols

df3 = df2.drop(columns=columns_to_drop)
df3
```

```python
# rename columns
df3 =  df3.rename(columns={'Month': 'year', 'Nominal': 'food_price_index'})
df4 = df3.copy()
df4
```

```python
# caluclate year-over-year % change

# convert 'year' and 'food_price_index' to numeric
df4['year'] = pd.to_numeric(df4['year'], errors='coerce')
df4['food_price_index'] = pd.to_numeric(df4['food_price_index'], errors='coerce')

# drop rows where 'year' or 'food_price_index' are NaN
df4.dropna(subset=['year', 'food_price_index'], inplace=True)

df4 = df4.sort_values('year')
df4['food_price_change_pct'] = df4['food_price_index'].pct_change() * 100

df4.head(10)
```

```python
df_food_index_clean = df4[(df4['year'] >= 1980) & (df4['year'] <= 2019)]

df_food_index_clean
```
