```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/Capstone/SIPRI-Milex-data-1949-2024-shareGDP.csv"

df = pd.read_csv(url)
df.head(15)
```

```python
# set the column names to the values of the sixth row

df.columns = df.iloc[4]
df.head(15)
```

```python
# drop the first 7 rows

df2 = df.drop(df.index[0:8])
df2.head(15)
```

```python
# check structure

print('Current Shape:', df2.shape)
print('Columns:', df2.columns.tolist())
```

```python
# drop columns we don't want

df3 = df2.drop(columns=['Notes','1949', '1950', '1951', '1952', '1953', '1954', '1955', '1956', '1957', '1958', '1959', '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979','2020', '2021', '2022', '2023', '2024'])
df3
```

```python
year_cols = [col for col in df3.columns if str(col).isdigit()]

# melt to long format
df_long = df3.melt(
    id_vars=['Country'],
    value_vars=year_cols,
    var_name='year',
    value_name='military_spending_pct_gdp'
)

df_long.head(20)
```
