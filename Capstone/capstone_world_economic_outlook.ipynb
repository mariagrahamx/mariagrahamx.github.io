```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/Capstone/World%20Economic%20Outlook%20Database%20-%201980%20to%202023.csv"

df = pd.read_csv(url)
df
```

```python
# check structure

print('Current Shape:', df.shape)
print('Columns:', df.columns.tolist())
```

```python
# drop columns we don't want

df2 = df.drop(columns=['WEO Country Code', 'WEO Subject Code', 'Country/Series-specific Notes', '2020','2021','2022','2022','2023','2024','2025','2026','2027','2028','2029','2030'])
df2
```

```python
# select indicators we want

variables_to_keep = {
    'Gross domestic product, constant prices': 'Percent change',
    'Gross domestic product per capita, constant prices': 'National currency',
    'Inflation, average consumer prices': 'Percent change',
    'Unemployment rate': 'Percent of total labor force',
    'Population':'Persons',
    'General government revenue': 'Percent of GDP',
    'General government net lending/borrowing': 'Percent of GDP',
    'Current account balance': 'Percent of GDP'
}

def filter_row(row):
    descriptor = row['Subject Descriptor']
    units = row['Units']

    if descriptor in variables_to_keep:
        expected_units = variables_to_keep[descriptor]
        return expected_units in units
    return False

df2_filtered = df2[df2.apply(filter_row, axis=1)].copy()

print(f"\nFiltered to {len(df2_filtered)} rows")
```

```python
df2_filtered.head(40)
```

```python
# clean names of indicators

variable_mapping = {
    'Gross domestic product, constant prices': 'gdp_growth',
    'Gross domestic product per capita, constant prices': 'gdp_pc_constant',
    'Inflation, average consumer prices': 'inflation_avg',
    'Unemployment rate': 'unemployment',
    'Population': 'population_millions',
    'General government revenue': 'gov_revenue',
    'General government net lending/borrowing': 'gov_fiscal_balance',
    'Current account balance': 'current_account_pct_gdp'
}

df2_filtered['variable_name'] = df2_filtered['Subject Descriptor'].map(variable_mapping)

df2_filtered
```

```python
# melt to long format

year_cols = [col for col in df2_filtered.columns if str(col).isdigit()]

print(f"Years: {min(year_cols)} to {max(year_cols)}")

df_long= df2_filtered.melt(
    id_vars=['ISO', 'Country', 'variable_name'],  # keep only these columns
    value_vars=year_cols,  # convert these to rows
    var_name='year',
    value_name='value'
)

# convert types

df_long['year'] = df_long['year'].astype(int)
df_long['value'] = pd.to_numeric(df_long['value'], errors='coerce')

print(f"\nLong format shape: {df_long.shape}")
```

```python
df_long.head(15)
```

```python
# pivot to wide format

df_wide = df_long.pivot_table(
    index=['ISO', 'Country', 'year'],  # one row per country-year
    columns='variable_name',            # each indicator becomes a column
    values='value',
    aggfunc='first'  # in case of any remaining duplicates
).reset_index()

print(f"\nWide format shape: {df_wide.shape}")
print(f"Columns: {df_wide.columns.tolist()}")
```
