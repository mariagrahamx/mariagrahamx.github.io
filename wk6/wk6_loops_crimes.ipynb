```python
import requests
import json
```

```python
data_url = "https://data.police.uk/api/crimes-street/crime-categories?date=2024-01&lat=52.629729&lng=-1.131592"
response = requests.get(data_url)
response.text
```

```python
target_series = [
    'anti-social-behaviour',
    'violent-crime',
    'bicycle-theft',
    'drugs',
    'burglary',
    'criminal-damage-arson'
]

base_url = "https://data.police.uk/api/crimes-street/{}?date=2024-01&lat=52.629729&lng=-1.131592"

for crime in target_series:
  # update base url
  new_url = base_url.format(crime)
  print(new_url)

  # after updating API url, make requests
  response = requests.get(new_url)

  # save the data on each iteration
  data = response.json()

  with open(f"data_{crime}.json", "w") as f:
    f.write(json.dumps(data))
```
