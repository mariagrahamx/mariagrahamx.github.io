```python
import pandas as pd
import altair as alt
```

```python
prices_df = pd.read_parquet('https://autocpi-public.s3.eu-west-2.amazonaws.com/lrpd/db_prices.parquet')
prices_df.describe().round()
```

```python
items_df = pd.read_parquet('https://autocpi-public.s3.eu-west-2.amazonaws.com/lrpd/db_item.parquet')
items_df.head(10)
```

```python
items_df[items_df['description'].str.contains("kebab", case=False)]
```

```python
kebab_prices = prices_df[prices_df['item_id'] == 220323 ].copy()
kebab_prices
```

```python
merge_df = pd.merge(kebab_prices, items_df, on='item_id', how='left')
merge_df
```

```python
region_df = pd.read_csv("https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/wk8/db_region.csv")
region_df.head(13)
```

```python
region_map = {
    2.0:"London",
    3.0: "South East",
    4.0: "South West",
    5.0: "East Anglia",
    6.0: "East Midlands",
    7.0:"West Mindlands",
    8.0: "Yorkshire & Humber",
    9.0: "North West",
    10.0: "North",
    11.0: "Wales",
    12.0: "Scotland",
    13.0: "Northern Ireland"
}
```

```python
df2 = merge_df.copy()
df2["region_name"] = df2["region"].map(region_map)
df2
```

```python
monthly_kebab_prices = df2.groupby(
    ["quote_date", "region_name", "description"]
).agg({"price": ["mean"]}).reset_index()

monthly_kebab_prices.columns = ["date", "region", "item", "mean"]

monthly_kebab_prices
```

```python
monthly_kebab_prices['date'] = pd.to_datetime(monthly_kebab_prices['date'], format='%Y%m')
monthly_kebab_prices
```

```python
monthly_kebab_prices.to_csv('wk8_kebab_prices.csv', index=False)
```
