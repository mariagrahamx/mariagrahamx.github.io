```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/project/simd2020_withinds.csv"

simd_full = pd.read_csv(url)

simd_full
```

```python
glasgow = simd_full[simd_full["Council_area"].str.strip().eq("Glasgow City")].copy()


cols = [
    "Data_Zone",
    "Intermediate_Zone",
    "Council_area",
    "SIMD2020v2_Income_Domain_Rank",
    "SIMD2020_Employment_Domain_Rank",
    "SIMD2020_Education_Domain_Rank",
    "SIMD2020_Health_Domain_Rank",
    "SIMD2020_Crime_Domain_Rank",
    "SIMD2020_Access_Domain_Rank"
]

# only copy columns i want
glasgow_clean = glasgow[cols].copy()

glasgow_clean
```

```python
# rename DZ -> Datazone for consistency
glasgow_clean = glasgow_clean.rename(columns={"Data_Zone": "Datazone", "SIMD2020v2_Income_Domain_Rank": "Income_Domain_Rank", "SIMD2020_Employment_Domain_Rank": "Employment_Domain_Rank",
                                        "SIMD2020_Education_Domain_Rank": "Education_Domain_Rank", "SIMD2020_Health_Domain_Rank":"Health_Domain_Rank", "SIMD2020_Crime_Domain_Rank": "Crime_Domain_Rank",
                                              "SIMD2020_Access_Domain_Rank":"Access_Domain_Rank" })

# make sure ranks are numeric
rank_cols = ["Income_Domain_Rank", "Employment_Domain_Rank", "Education_Domain_Rank", "Health_Domain_Rank", "Crime_Domain_Rank", "Access_Domain_Rank"]
glasgow_clean[rank_cols] = glasgow_clean[rank_cols].apply(
    pd.to_numeric, errors="coerce"
)

# drop rows with missing Datazone or all ranks missing
glasgow_simd_clean = glasgow_clean.dropna(
    subset=["Datazone"], how="any"
)

glasgow_simd_clean
```

```python
glasgow_simd_clean.to_csv("glasgow_SIMD_clean.csv", index=False)
```

```python
import geopandas as gpd
import pandas as pd
import warnings; warnings.filterwarnings('ignore', 'GeoSeries.notna', UserWarning)
```

```python
url_2 = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/project/scotland_datazones_2011.json"

gdf = gpd.read_file(url_2)

gdf
```

```python
url_3 = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/project/glasgow_SIMD_clean.csv"

simd = pd.read_csv(url_3)

simd
```

```python
url_4 = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/project/glasgow_derelict_sites_clean.csv"

sites = pd.read_csv(url_4)

sites
```

```python
# keep only polygons that appear in Glasgow SIMD
gdf = gdf[gdf["DataZone"].isin(simd["Datazone"])].copy()
print("Polygons kept:", len(gdf))
```

```python
print(gdf["DataZone"].head(5).tolist())
print(gdf["DataZone"].tail(5).tolist())
print(gdf["DataZone"].nunique())
```

```python
print(simd["Datazone"].head(5).tolist())
print(simd["Datazone"].tail(5).tolist())
print(simd["Datazone"].nunique())
```

```python
print(sites["Datazone"].head(5).tolist())
print(sites["Datazone"].tail(5).tolist())
print(sites["Datazone"].nunique())
```

```python
sites_gdf = gpd.GeoDataFrame(
    sites,
    geometry=gpd.points_from_xy(sites["East"], sites["North"]),
    crs="EPSG:27700"
)

sites_gdf
```

```python
print(gdf.crs)
```

```python
# set the initial CRS
gdf.set_crs("EPSG:4326", inplace=True)

# british national grid
gdf_dz = gdf.to_crs(epsg=27700)

print(gdf_dz.geometry.iloc[0])
```

```python
gdf_dz["centroid"] = gdf_dz.geometry.centroid
dz_centroids = gpd.GeoDataFrame(
    gdf_dz[["DataZone", "centroid"]],
    geometry="centroid",
    crs="EPSG:27700"
)

dz_centroids
```

```python
# for each datazone centroid find nearest site distance
dz_centroids["dist_to_site_m"] = dz_centroids.geometry.apply(
    lambda x: sites_gdf.distance(x).min()
)

dz_centroids["dist_km"] = dz_centroids["dist_to_site_m"] / 1000
```

```python
simd_with_dist = simd.merge(
    dz_centroids[["DataZone", "dist_km"]],
    left_on="Datazone",
    right_on="DataZone",
    how="left"
)

simd_with_dist.head()
```

```python
overlap = set(simd["Datazone"].astype(str)).intersection(set(dz_centroids["DataZone"].astype(str)))
len(overlap), list(sorted(overlap))[:5]
```

```python
print("NaN share:", simd_with_dist["dist_km"].isna().mean())
print("Rows with distance:", simd_with_dist["dist_km"].notna().sum(), "/", len(simd_with_dist))
```

```python
wide =simd_with_dist[[
    "Datazone", "dist_km",
    "Income_Domain_Rank", "Employment_Domain_Rank", "Education_Domain_Rank", "Health_Domain_Rank", "Crime_Domain_Rank",
    "Access_Domain_Rank"
]].copy()

wide
```

```python
wide.to_csv("glasgow_simd_2020_distance_wide.csv", index=False)
```

```python
long = simd_with_dist.melt(
    id_vars=["Datazone", "dist_km"],
    value_vars=[
        "Income_Domain_Rank",
        "Employment_Domain_Rank",
        "Education_Domain_Rank",
        "Health_Domain_Rank",
        "Crime_Domain_Rank",
        "Access_Domain_Rank"
    ],
    var_name="indicator",
    value_name="rank"
)

long
```

```python
long.to_csv("glasgow_simd_2020_distance_long.csv", index=False)
```
