```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/project/simd2020_withinds.csv"

simd_full = pd.read_csv(url)

simd_full
```

```python
glasgow = simd_full[simd_full["Council_area"].str.strip().eq("Glasgow City")].copy()


cols = [
    "Data_Zone",
    "Intermediate_Zone",
    "Council_area",
    "SIMD2020v2_Income_Domain_Rank",
    "SIMD2020_Employment_Domain_Rank",
    "SIMD2020_Education_Domain_Rank",
    "SIMD2020_Health_Domain_Rank",
    "SIMD2020_Crime_Domain_Rank"
]

# only copy columns i want
glasgow_clean = glasgow[cols].copy()

glasgow_clean
```

```python
# rename DZ -> Datazone for consistency
glasgow_clean = glasgow_clean.rename(columns={"Data_Zone": "Datazone", "SIMD2020v2_Income_Domain_Rank": "Income_Domain", "SIMD2020_Employment_Domain_Rank": "Employment_Domain",
                                        "SIMD2020_Education_Domain_Rank": "Education_Domain", "SIMD2020_Health_Domain_Rank":"Health_Domain", "SIMD2020_Crime_Domain_Rank": "Crime_Domain" })

# make sure ranks are numeric
rank_cols = ["Income_Domain", "Employment_Domain", "Education_Domain", "Health_Domain", "Crime_Domain"]
glasgow_clean[rank_cols] = glasgow_clean[rank_cols].apply(
    pd.to_numeric, errors="coerce"
)

# drop rows with missing Datazone or all ranks missing
glasgow_simd_clean = glasgow_clean.dropna(
    subset=["Datazone"], how="any"
)

glasgow_simd_clean
```

```python
glasgow_simd_clean.to_csv("glasgow_SIMD_clean.csv", index=False)
```
