```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/mariagrahamx/mariagrahamx.github.io/refs/heads/main/wk3/Data_for_Neonatal_Deaths_By_Ethnicity_%20UK.csv"
data1 = pd.read_csv(url)
data1
```

```python
data2 = data1[data1['Model'].str.contains('Model 2', na=False)].copy()
data2
```

```python
lookup = {
    'Asian/Asian British: Bangladeshi': 'Asian British: Bangladeshi',
    'Asian/Asian British: Indian': 'Asian British: Indian',
    'Asian/Asian British: Pakistani': 'Asian British: Pakistani',
    'Black/African/Caribbean/Black British: African': 'Black British: African',
    'Black/African/Caribbean/Black British: Caribbean': 'Black British: Caribbean',
    'Mixed/multiple ethnic groups': "Mixed/Multiple Ethnic Groups",
    'Other ethnic group': 'Other Ethnic Group'
    }

data2['Group_short'] = data2['Group'].replace(lookup)
```

```python
data2.to_csv('wk3_neonatal_mortality_data_2011_2017.csv', index=False)
```
