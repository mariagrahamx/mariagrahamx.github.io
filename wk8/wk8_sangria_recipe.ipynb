```python
sangria_recipe = {
    310421: "Red wine",
    212709: "Oranges",
    212717: "Apples",
    212008: "Lemonade",
    310405: "Brandy",
    212101: "Sugar"
}
```

```python
sangria_recipe_items = items_df[items_df['item_id'].isin(sangria_recipe)]
sangria_recipe_items
```

```python
common_start = sangria_recipe_items["date_quote_s"].max()
common_end   = sangria_recipe_items["date_quote_e"].min()

common_start, common_end
```

```python
sangria_recipe_items[
    (sangria_recipe_items["date_quote_s"] <= common_start) &
    (sangria_recipe_items["date_quote_e"] >= common_end)
]
```

```python
sangria_df = (prices_df.merge(sangria_recipe_items[["item_id", "description"]], how="inner",on="item_id"))
sangria_df
```

```python
sangria_df["quote_date"] = sangria_df["quote_date"].astype(int)

sangria_df2 = sangria_df[
    (sangria_df["quote_date"] >= common_start) &
    (sangria_df["quote_date"] <= common_end)
]
```

```python
monthly_sangria_prices = sangria_df2.groupby(['quote_date', 'description', 'item_id']).agg({'price': ['mean']}).reset_index()
monthly_sangria_prices
```

```python
monthly_sangria_prices['item'] = monthly_sangria_prices['item_id'].map(sangria_recipe)
monthly_sangria_prices
```

```python
monthly_sangria_prices.columns = ['date', 'description', 'item id', 'mean price', 'item'] # manually overwriting column names
monthly_sangria_prices
```

```python
monthly_sangria_prices['date'] = pd.to_datetime(monthly_sangria_prices['date'], format='%Y%m')
monthly_sangria_prices
```

```python
monthly_sangria_prices.to_csv('wk8_sangria_recipe.csv', index=False)
```
