```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_excel('/content/drive/MyDrive/Data for Maternal Mortality Rates by Ethnic Group.xlsx')
data1
```

```python
# set the column names to the values of the first row
data1.columns = data1.iloc[0]
data1
```

```python
# drop the first row now that it has been set as the column names
data2 = data1.drop(data1.index[0])
data2
```

```python
# list column names
data2.columns
```

```python
# delete CI columns
data3 = data2.drop(columns=['Lower CI', 'Upper CI'])
data3
```

```python
# rename Chinese/\nother column
data3.rename(columns={'Chinese/\nother': 'Chinese/Other'}, inplace=True)
data3.columns
```

```python
import re
def mid_from_range(s):
    m = re.search(r'(\d{4})\s*[-]\s*(\d{2,4})', str(s))
    if not m:
        return pd.NA
    start = int(m.group(1))
    end_raw = m.group(2)
    end = int(end_raw) if len(end_raw) == 4 else int(str(start)[:2] + end_raw)  # e.g., '11' -> 2011
    return start + (end - start) / 2

data3['mid_year'] = data3['Years'].apply(mid_from_range).astype(float)
data3['mid_year_int'] = data3['mid_year'].round().astype('Int64')
data3
```

```python
data3.rename(columns={'mid_year_int': 'Mid-Year'}, inplace=True)
data3.columns
```

```python
data4 = data3.drop(columns=['mid_year'])
data4
```

```python
# columns representing ethnic groups (used for melting the dataframe into long format)

eth_cols = ['White','Black','Asian','Chinese/Other','Mixed']

# convert from wide to long format
long = data4.melt(
    id_vars=['Years', 'Mid-Year'],
    value_vars=eth_cols,
    var_name='Ethnicity',
    value_name='Rate'   # maternal deaths per 100,000
)

# make sure Mid-Year is plain int to avoid warnings
long['Mid-Year'] = long['Mid-Year'].astype(int)

long
```

```python
# drop empty rows in Rate column

long2 = long.dropna(subset=['Rate'])
```

```python
latest = long2
latest
```

```python
latest.to_csv('wk3_maternal_mortality_data_2009_2023.csv', index=False)
```
